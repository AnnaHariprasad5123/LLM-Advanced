### Handling Null values

#### 1.  Mean, Median and Mode Imputation

```python
import pandas as pd
import numpy as np
```

```python
df_titanic = pd.read_csv("../../datasets/Titanic-Dataset.csv")
df_titanic.head()
```

```python
df_titanic.info()
```

```python
df_titanic.isnull().sum()
```

##### Genearalized Imputation - mean, median, and mode
- fillna()


```python
df_titanic['Age'].median()
```

```python
df_titanic['Age'].fillna(df_titanic['Age'].median(),inplace=True)
```

```python
df_titanic['Embarked'].mode()
```

```python
df_titanic['Embarked'].fillna(df_titanic['Embarked'].mode()[0],inplace=True)
```

```python
df_titanic.info()
```

- replace()

```python
df_titanic = pd.read_csv("../../datasets/Titanic-Dataset.csv")
df_titanic.head()
```

```python
df_titanic.isnull().sum()
```

```python
df_titanic['Age'].mean()
```

```python
df_titanic['Age'].replace(np.nan, df_titanic['Age'].mean(), inplace=True)
```

```python
df_titanic['Embarked'].replace(np.nan, df_titanic['Embarked'].mode()[0], inplace=True)
```

```python
df_titanic.isnull().sum()
```

### 2. Constant Value Imputation

#### 3. Forward and Backward Fill
- ffill()
- bfill()

```python
df_titanic = pd.read_csv("../../datasets/Titanic-Dataset.csv")
```

```python
df_titanic.isnull().sum()
```

```python
df_titanic[df_titanic['Age'].isnull()==True]
```

```python
df_titanic['Age'].ffill(inplace=True)
```

```python
df_titanic.head(20)
```

```python
df_titanic[df_titanic['Embarked'].isnull()==True]
```

```python
df_titanic['Embarked'].bfill(inplace=True)
```

```python
df_titanic.iloc[60:63,:]
```

#### 4. Indicator for Missingness (Flagging)

```python
df_titanic = pd.read_csv("../../datasets/Titanic-Dataset.csv")
```

```python
df_titanic.isnull().sum()
```

```python
df_titanic['Embarked_Null_Indicator'] = np.where(df_titanic['Embarked'].isnull()==True,-1,1) # -1 means null and 1 means not null
```

```python
df_titanic[df_titanic['Embarked'].isnull() == True]
```
